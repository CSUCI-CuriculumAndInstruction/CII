```
#@title Curriculum Ideologies Inventory { run: "auto", vertical-output: true, display-mode: "form" }
from IPython.display import clear_output
import pandas as pd
import textwrap
from matplotlib import pyplot as plt

# Part 1 Purpose
part1_A = "A. Schools should provide children with the ability to perceive problems in society, envision a better society, and act to change society so that there is social justice and a better life for all people."
part1_B = "B. Schools should fulfill the needs of society by efficiently training youth to function as mature constructive members of society."
part1_C = "C. Schools should be communities where the accumulated knowledge of the culture is transmitted to the youth."
part1_D = "D. Schools should be enjoyable, stimulating, child-centered environments organized around the developmental needs and interests of children as those needs and interests present themselves from day to day."

# Part 2 Teaching
part2_A = "A. Teachers should be supervisors of student learning, utilizing instructional strategies that will optimize student learning."
part2_B = "B. Teachers should be companions to students, using the environment within which the student lives to help the student learn."
part2_C = "C. Teachers should be aids to children, helping them learn by presenting them with experiences from which they can make meaning."
part2_D = "D. Teachers should be knowledgeable people, transmitting that which is known to those who do not know it."

# Part 3 Learning
part3_A = "A. Learning best proceeds when the student is presented with the appropriate stimulus materials and positive reinforcement."
part3_B = "B. Learning best proceeds when the teacher clearly and accurately presents to the student that knowledge which the student is to acquire."
part3_C = "C. Learning best takes place when children are motivated to actively engage in experiences that allow them to create their own knowledge and understanding of the world in which they live."
part3_D = "D. Learning best occurs when a student confronts a real social crisis and participates in the construction of a solution to that crisis."

# Part 4 Knowledge
part4_A = "A. The knowledge of most worth is the structured knowledge and ways of thinking that have come to be valued by the culture over time."
part4_B = "B. The knowledge of most worth is the personal meaning of oneself and of one’s world that comes from one’s direct experience in the world and one’s personal response to such experience."
part4_C = "C. The knowledge of most worth is the specific skills and capabilities for action that allow an individual to live a constructive life."
part4_D = "D. The knowledge of most worth is a set of social ideals, a commitment to those ideals, and an understanding of how to implement those ideals."

# Part 5 Childhood
part5_A = "A. Childhood is essentially a time of learning in preparation for adulthood, when one will be a constructive, contributing member of society."
part5_B = "B. Childhood is essentially a period of intellectual development highlighted by growing reasoning ability and capacity for memory that results in ever greater absorption of cultural knowledge."
part5_C = "C. Childhood is essentially a time when children unfold according to their own innate natures, felt needs, organic impulses, and internal timetables. The focus is on children as they are during childhood rather than as they might be as adults."
part5_D = "D. Childhood is essentially a time for practice in and preparation for acting upon society to improve both oneself and the nature of society."

# Part 6 Evaluation
part6_A = "A. Evaluation should objectively indicate to others whether or not students can or cannot perform specific skills. Its purpose is to certify students’ competence to perform specific tasks."
part6_B = "B. Evaluation should continuously diagnose children’s needs and growth so that further growth can be promoted by appropriate adjustment of their learning environment. It is primarily for the children’s benefit, not for comparing children with each other or measuring them against predetermined standards."
part6_C = "C. Evaluation should be a subjective comparison of students’ performance with their capabilities. Its purpose is to indicate to both the students and others the extent to which they are living up to their capabilities."
part6_D = "D. Evaluation should objectively determine the amount of knowledge students have acquired. It allows students to be ranked from those with the greatest intellectual gain to those with the least."

intro_txt = "Read each of the following parts carefully. Then, rank each statement in order from most liked to least liked. This is not a test and there is no one right answer. Take your time. Press enter to move to the next section."
rank_input = "Enter the letter value for each statement ranked from most liked to least liked:\n"

wrapper = textwrap.TextWrapper(width=80)

parts = [
    [part1_A, part1_B, part1_C, part1_D],
    [part2_A, part2_B, part2_C, part2_D],
    [part3_A, part3_B, part3_C, part3_D],
    [part4_A, part4_B, part4_C, part4_D],
    [part5_A, part5_B, part5_C, part5_D],
    [part6_A, part6_B, part6_C, part6_D],
]


def part(n):
    print()
    print("\033[1m" + f"PART {n}" + "\033[0m")
    print()

    for statement in parts[n - 1]:
        print(wrapper.fill(text=statement), end=(2 * "\n"))

def rank(answer):
    answer = answer[::-1].replace(" ", "").replace(",", "").upper()
    return answer

df = pd.DataFrame(columns=['Part','Scholar Academic', 'Learner Centered', 'Social Efficiency', 'Social Reconstruction'])


scholar_academic = 0
learner_centered = 0
social_reconstruction = 0
social_efficiency = 0

print(wrapper.fill(text=intro_txt))
input()
clear_output() 

part(1)
print(wrapper.fill(text=rank_input))
part1_rank = rank(input())
df.at['0', 'Part'] = 'Purpose'
for value, statement in enumerate(part1_rank):
    if statement == 'A':
        social_reconstruction += (value + 1)
        df.at['0','Social Reconstruction'] = value+1    
    elif statement == 'B':
        social_efficiency += (value + 1)
        df.at['0','Social Efficiency'] = value+1 
    elif statement == 'C':
        scholar_academic += (value + 1)
        df.at['0','Scholar Academic'] = value+1 
    elif statement == 'D':
        learner_centered += (value + 1)
        df.at['0','Learner Centered'] = value+1 
clear_output() 

part(2)
print(wrapper.fill(text=rank_input))
part2_rank = rank(input())
df.at['1', 'Part'] = 'Teaching'

for value, statement in enumerate(part2_rank):
    if statement == 'A':
        social_efficiency += (value + 1)
        df.at['1','Social Efficiency'] = value+1 
    elif statement == 'B':
        social_reconstruction += (value + 1)
        df.at['1','Social Reconstruction'] = value+1 
    elif statement == 'C':
        learner_centered += (value + 1)
        df.at['1','Learner Centered'] = value+1 
    elif statement == 'D':
        scholar_academic += (value + 1)
        df.at['1','Scholar Academic'] = value+1 
clear_output() 

part(3)
print(wrapper.fill(text=rank_input))
part3_rank = rank(input())
df.at['2', 'Part'] = 'Learning'
for value, statement in enumerate(part3_rank):
    if statement == 'A':
        social_efficiency += (value + 1)
        df.at['2','Social Efficiency'] = value+1 
    elif statement == 'B':
        scholar_academic += (value + 1)
        df.at['2','Scholar Academic'] = value+1 
    elif statement == 'C':
        learner_centered += (value + 1)
        df.at['2','Learner Centered'] = value+1 
    elif statement == 'D':
        social_reconstruction += (value + 1)
        df.at['2','Social Reconstruction'] = value+1 
clear_output() 

part(4)
print(wrapper.fill(text=rank_input))
part4_rank = rank(input())
df.at['3', 'Part'] = 'Knowledge'
for value, statement in enumerate(part4_rank):
    if statement == 'A':
        scholar_academic += (value + 1)
        df.at['3','Scholar Academic'] = value+1 
    elif statement == 'B':
        learner_centered += (value + 1)
        df.at['3','Learner Centered'] = value+1 
    elif statement == 'C':
        social_efficiency += (value + 1)
        df.at['3','Social Efficiency'] = value+1 
    elif statement == 'D':
        social_reconstruction += (value + 1)
        df.at['3','Social Reconstruction'] = value+1 
clear_output() 

part(5)
print(wrapper.fill(text=rank_input))
part5_rank = rank(input())
df.at['4', 'Part'] = 'Childhood'
for value, statement in enumerate(part5_rank):
    if statement == 'A':
        social_efficiency += (value + 1)
        df.at['4','Social Efficiency'] = value+1 
    elif statement == 'B':
        scholar_academic += (value + 1)
        df.at['4','Scholar Academic'] = value+1 
    elif statement == 'C':
        learner_centered += (value + 1)
        df.at['4','Learner Centered'] = value+1 
    elif statement == 'D':
        social_reconstruction += (value + 1)
        df.at['4','Social Reconstruction'] = value+1 
clear_output()

part(6)
print(wrapper.fill(text=rank_input))
part6_rank = rank(input())
df.at['5', 'Part'] = 'Evaluation'
for value, statement in enumerate(part6_rank):
    if statement == 'A':
        social_efficiency += (value + 1)
        df.at['5','Social Efficiency'] = value+1 
    elif statement == 'B':
        learner_centered += (value + 1)
        df.at['5','Learner Centered'] = value+1 
    elif statement == 'C':
        social_reconstruction += (value + 1)
        df.at['5','Social Reconstruction'] = value+1 
    elif statement == 'D':
        scholar_academic += (value + 1)
        df.at['5','Scholar Academic'] = value+1 
clear_output() 

print()
print("RESULTS")
print("-------")
print(f"Scholar_Academic = {scholar_academic/6: .2f}")
print(f"Learner Centered = {learner_centered/6: .2f}")
print(f"Social Efficiency = {social_efficiency/6: .2f}")
print(f"Social Reconstruction = {social_reconstruction/6: .2f}")
print()

x = df['Part']

curriculums = ['Scholar Academic', 'Learner Centered', 'Social Efficiency', 'Social Reconstruction']

for i, ideology in enumerate(curriculums):
    plt.style.use('fivethirtyeight')
    f = plt.figure(i+1)
    f.set_figwidth(6.6)
    f.set_figheight(2.25)
    plt.ylim(0.5, 4.5)
    plt.title(ideology)
    plt.plot(x, df[ideology])
    plt.ylabel('Score')

plt.show()
```
